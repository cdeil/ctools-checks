```python
import gammalib
```

```python
rng = gammalib.GRan(0)
```

```python
size = 1e7
vals = [rng.uniform() for _ in range(int(size))]
```

```python
import numpy as np
print(np.min(vals), np.max(vals))
```

```python
%pylab inline --no-import-all
import matplotlib.pyplot as plt
```

```python
plt.hist2d(vals[:len(vals)/2], vals[len(vals)/2:], bins=47);
plt.colorbar();
```
