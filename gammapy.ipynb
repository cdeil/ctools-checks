```python
from gammapy.spectrum import crab

for ref in crab.REFERENCES:
    emin, emax = 1, 10 # TeV
    int_flux = crab.crab_integral_flux(emin, emax, ref) # cm^-1 s
    print('%15s %.3g' % (ref, int_flux))
```

```python
import numpy as np
from astropy.units import Unit
import matplotlib.pyplot as plt

energy = np.logspace(-2, 2, 100) # TeV

for ref in crab.REFERENCES:
    flux = crab.crab_flux(energy, ref) # cm^-2 s^-1 TeV^-1
    flux_sed = Unit('TeV').to('erg') * energy ** 2 * flux
    plt.plot(energy, flux_sed, label=ref)

plt.xlabel('Energy (TeV)')
plt.ylabel('Flux (erg cm^-2 s^-1)')
plt.ylim(1e-12, 1e-9)
plt.loglog()
plt.legend();
```
